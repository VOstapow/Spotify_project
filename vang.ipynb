```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Csv pathway 
songs_path = "../final.csv"

# Read csv file
df = pd.read_csv(songs_path, low_memory=False) 
df.head()
```

```python
# Create new DataFrame 
new_df = pd.DataFrame(df[["uri", "rank", "artist_names", "artist_id", "artist_img", "track_name", "album_cover", "peak_rank", "weeks_on_chart", "streams", "week", "country", "danceability", "acousticness", "loudness", "tempo"]])
new_df = new_df.loc[new_df["rank"]!="rank"]
# new_df.set_index("uri", inplace=True)
new_df.head()
```

```python
new_df.dtypes
```

```python
# Convert datatypes
types ={"rank": int, 
        "peak_rank": int,
        "weeks_on_chart": int, 
        "streams": int, 
        "week": "datetime64", 
        "danceability": float, 
        "acousticness": float, 
        "loudness": float, 
        "tempo": float 
        }
new_df = new_df.astype(types)
new_df.dtypes
```
