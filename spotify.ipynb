```python
# Import
import pandas as pd 

# csv path 
songs_path = "Resources/final.csv"

# Read csv file
df = pd.read_csv(songs_path, low_memory=False)
df.head()
```

```python
# Create new DataFrame
spotify_df = pd.DataFrame(df[["uri", "rank", "artist_names", "artist_id", "artist_genre", "artist_img", "track_name", "album_cover", "peak_rank", "weeks_on_chart", "streams", "week", "country", "danceability", "acousticness", "loudness", "tempo"]])
spotify_df = spotify_df.loc[spotify_df["rank"]!="rank"]
spotify_df["rank"].unique()
spotify_df.head()
```

```python
# Convert DataTypes
types ={"rank": int, 
        "peak_rank": int,
        "weeks_on_chart": int, 
        "streams": int, 
        "week": "datetime64", 
        "danceability": float, 
        "acousticness": float, 
        "loudness": float, 
        "tempo": float 
        }
spotify_df = spotify_df.astype(types)
spotify_df.dtypes
```

```python
# Narrow DataFrame
spotify_df = spotify_df.loc[(spotify_df["country"] == "United States")
                            | (spotify_df["country"] == "Canada")
                            | (spotify_df["country"] == "Mexico")
                            & (spotify_df["week"] >= "06-14-2021")]
spotify_df
```
