```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# Study data files
songs_path = "final.csv"

# Read the mouse data and the study results
songs_db = pd.read_csv(songs_path, low_memory=False)

songs_db.head()
```

```python
del songs_db['Unnamed: 0']
```

```python
songs_db['artists_num'].unique()
```

```python
df=songs_db.loc[songs_db['artists_num']!='artists_num']
df.head()
```

```python
#Convert 'rank' to numeric
df['rank']=pd.to_numeric(df['rank'],errors='coerce')
```

```python
#Convert 'peak_rank' to numeric
df['peak_rank']=pd.to_numeric(df['peak_rank'],errors='coerce')
```

```python
#Convert 'weeks_on_chart' to numeric
df['weeks_on_chart']=pd.to_numeric(df['weeks_on_chart'],errors='coerce')
```

```python
#Convert 'streams' to numeric
df['streams']=pd.to_numeric(df['streams'],errors='coerce')
```

```python
#Convert 'danceability' to numeric
df['danceability']=pd.to_numeric(df['danceability'],errors='coerce')
```

```python
#Convert 'acousticness' to numeric
df['acousticness']=pd.to_numeric(df['acousticness'],errors='coerce')
```

```python
#Convert 'loudness' to numeric
df['loudness']=pd.to_numeric(df['loudness'],errors='coerce')
```

```python
#Convert 'tempo' to numeric
df['tempo']=pd.to_numeric(df['tempo'],errors='coerce')
```

```python
df.info()
```

```python
df = df[["uri","rank","artist_names","artist_id","artist_genre","artist_img","track_name","album_cover",\
   "peak_rank","weeks_on_chart","streams","week","danceability","acousticness","loudness","tempo","country"]]
```
